```python
from googletrans import Translator,LANGUAGES

class Mytranslator:
    def __init__(self):
        self.langs=list(LANGUAGES.values())
    def run(self,txt='Type your text here',src='English',dest='Hindi'):
        self.translator=Translator()
        self.txt=txt
        self.src=src
        self.dest=dest
        try:
            self.translated=self.translator.translate(self.txt,src=self.src,dest=self.dest)
        except:
            self.translated=self.translator.translate(self.txt)
        self.ttext=self.translated.text
        return self.ttext
    
    
import tkinter as tk
from tkinter import ttk

root=tk.Tk()
root.title('Translator')
root.geometry('900x500')
root.config(bg='Black')

def get():
    s=srclan.get()
    d=destlan.get()
    message=sourcetext.get(1.0,'end')
    translator=Mytranslator()
    text=translator.run(src=s,dest=d,txt=message)
    desttext.delete(1.0,'end')
    desttext.insert('end',text)

l=tk.Label(root,text='Language Translator', bd=5,relief='sunken')
l.config(bg='#FFFFFF',fg='#000000', font=('Courier', 20, 'bold'))
l.place(relx=0.5, rely=0.02, relwidth=0.5, relheight=0.1, anchor='n')

frame=tk.Frame(root).pack(side=tk.BOTTOM)

langs=Mytranslator().langs

l1 = tk.Label(frame, text="Select Source Language",bd=2,relief='sunken')
l1.config(bg='#000000', fg='#ffffff', font=('Courier', 12, 'bold'))
l1.place(relx=0.05, rely=0.2, relwidth=0.28, relheight=0.07)

srclan=ttk.Combobox(frame,values=langs,width=10)
srclan.place(relx=0.33, rely=0.2, relwidth=0.13, relheight=0.07)
srclan.set('english')

l2 = tk.Label(frame, text="Select Destination Language",bd=2,relief='sunken')
l2.config(bg='#000000', fg='#ffffff', font=('Courier', 12, 'bold'))
l2.place(relx=0.52, rely=0.2, relwidth=0.31, relheight=0.07)

destlan=ttk.Combobox(frame,values=langs,width=10)
destlan.place(relx=0.83, rely=0.2, relwidth=0.11, relheight=0.07)
destlan.set('hindi')

sc1=tk.Scrollbar(frame)
sourcetext=tk.Text(frame,font=('arial',10),height=11,wrap=tk.WORD)

sc1.place(relx=0.46, rely=0.3, relwidth=0.02, relheight=0.4, anchor='n')
sourcetext.place(relx=0.25, rely=0.3, relwidth=0.4, relheight=0.4, anchor='n')

sc2=tk.Scrollbar(frame)
desttext=tk.Text(frame,font=('arial',10),height=11,wrap=tk.WORD)

desttext.place(relx=0.72, rely=0.3, relwidth=0.4, relheight=0.4, anchor='n')
sc2.place(relx=0.93, rely=0.3, relwidth=0.02, relheight=0.4, anchor='n')

btn=tk.Button(frame,text='Translate',command=lambda : get())
btn.config(bg='#eeeeee', fg='black', font=('Courier', 15, 'bold'),border=5)
btn.place(relx=0.45, rely=0.8, relwidth=0.17, relheight=0.06)

root.mainloop()
```
